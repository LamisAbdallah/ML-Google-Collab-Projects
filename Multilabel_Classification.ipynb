```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
train_data = pd.read_csv("/kaggle/input/multilabel-classification-dataset/train.csv")
X_train = train_data['ABSTRACT'].values
y_train = train_data[['Computer Science', 'Physics', 'Mathematics', 'Statistics', 'Quantitative Biology', 'Quantitative Finance']].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Text preprocessing
vectorizer = TfidfVectorizer(max_features=1000)
X_train_vec = vectorizer.fit_transform(X_train).toarray()
X_test_vec = vectorizer.transform(X_test).toarray()
```

```python
# Normalize the input data
X_train_vec = X_train_vec / np.max(X_train_vec)
X_test_vec = X_test_vec / np.max(X_test_vec)
```

```python
# sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Build the Keras model
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_vec.shape[1],)),  # Larger hidden layer
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(y_train.shape[1], activation='sigmoid')  # Output layer
])
```

```python
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train_vec, y_train, epochs=50, batch_size=32, validation_data=(X_test_vec, y_test))
```
